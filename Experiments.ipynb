## Experiments with the gains of companies

```python
# List of 48 companies 
tickers = ["VTRS", "UHS", "SYK", "LH", "RMD", "PFE", "MCK", "MRK", "ABT", "JNJ", "ISRG", "HSIC", "EW", "DVA", "DHR", "COO", "IPG", "OMC", "DE", "ADM", "BG", 
           "AMP", "BLK", "BK", "BEN", "IVZ", "KKR", "F", "GM", "TSLA", "BWA", "AZO", "KMX", "RCL", "HST", "RCL", "BKNG", "CCL", "EXPE", 
            "CVX", "XOM", "HES", "BKR", "HAL", "SLB", "APA", "COP", "CTRA", "DVN"]

data = yf.download(tickers, start="2017-01-01", end="2017-12-31")['Close']

# Compute daily gains
returns = data.pct_change().dropna()

returns
```

```python
clustering(tickers, returns.T, nb_clusters = 2, year = '2017', THRESHOLD=0.2).plot_pie_for_each_cluster()
```
